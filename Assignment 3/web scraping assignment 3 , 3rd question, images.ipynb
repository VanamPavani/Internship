```python
#print(ProductName)

productNames = ["fruits", "Cars" ,"Machine Learning", "Guitar", "Cakes"]
```

```python
img_urls_fruits = []
img_urls_Cars = []
img_urls_Learning = []
img_urls_Guitar = []
img_urls_Cakes = []

for i in productNames :
    driver.get('https://images.google.com/')
    search_field=driver.find_element_by_xpath("//input[@class='gLFyf gsfi']")
    print(search_field.text)    

    search_field.send_keys(i)
    search_button=driver.find_element_by_xpath("//button[@class='Tg7LZd']")
    driver.execute_script("arguments[0].click();", search_button)
    imgResults = driver.find_elements_by_xpath("//img[contains(@class,'rg_i Q4LuWd')]")
    
    for j in  range(0,10):
        img=imgResults[j]
        try:
            img.click()
            time.sleep(2)
            actual_images = driver.find_elements_by_css_selector('img.rg_i.Q4LuWd')
            count = 0
            for actual_image in actual_images:
                if actual_image.get_attribute('src') and 'https' in actual_image.get_attribute('src'):
                    if i == 'fruits' :
                        img_urls_fruits.append(actual_image.get_attribute('src'))
                    if i == 'Cars' :
                        img_urls_Cars.append(actual_image.get_attribute('src'))
                    if i == 'Machine Learning' :
                        img_urls_Learning.append(actual_image.get_attribute('src'))
                    if i == 'Guitar' :
                        img_urls_Guitar.append(actual_image.get_attribute('src'))
                    if i == 'Cakes' :
                        img_urls_Cakes.append(actual_image.get_attribute('src'))
                    count=count+1
                if count > 9 :
                    break
        except ElementClickInterceptedException or ElementNotInteractableException as err:
            print(err)
print(img_urls_fruits)
print(len(img_urls_fruits))
print(img_urls_Cars)
print(len(img_urls_Cars))
print(img_urls_Learning)
print(len(img_urls_Learning))
print(img_urls_Guitar)
print(len(img_urls_Guitar))
print(img_urls_Cakes)
print(len(img_urls_Cakes))
#driver.find_element_by_xpath("//button[@class='M6CB1c rr4y5c']").click()
```
