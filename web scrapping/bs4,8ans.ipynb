```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_title = soup.find('span', class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full")
first_title

```

```python
first_titles = []
for i in soup.find_all('span', class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"):
    first_titles.append(i.text)
first_titles
```

```python
location = soup.find('div', class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95")
location
```
